```python
features = ["Age", "Income", "Total_Spending", "NumWebPurchases", "NumStorePurchases", "NumWebVisitsMonth", "Recency" ]
```

```python
x = df[features].copy()
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,10), wcss,marker = "o")
plt.title("Eblow Method for Optiomal K")
plt.xlabel("Number of Clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters=6)
df["Clusters"] = kmeans.fit_predict(x_scaled)
```

```python
df["Clusters"]
```

```python
# Step 1: Define the feature columns
features = ['Age', 'Income', 'Total_Spending', 'NumWebPurchases',
            'NumStorePurchases', 'NumWebVisitsMonth', 'Recency']

```

```python
# Step 2: Create the cluster summary
clusters_summary = df.groupby("Clusters")[features].mean().round(1)
clusters_summary["Count"] = df.groupby("Clusters").size()

```

```python
# Step 3: Transpose for plotting (excluding "Count" column)
clusters_summary_no_count = clusters_summary.drop("Count", axis=1).T


```

```python

clusters_summary_no_count.plot(kind='bar', figsize=(12, 6))
plt.title("Cluster Summary by Feature")
plt.xlabel("Features")
plt.ylabel("Average Value")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()
```

```python
clusters_summary.to_csv("cluster_summary.csv")
```
